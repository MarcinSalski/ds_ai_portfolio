```python
import pandas as pd
```

# Najpierw obejrzę całą bazę

```python
df = pd.read_csv('27__dane.csv', sep=';')

df
```

# Duplikaty

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

# Imię

```python
df['Imię'] = df['Name'].apply(lambda x: x.split(' ')[0] if pd.notnull(x) else None)
df
```

# Nazwisko

```python
df['Nazwisko'] = df['Name'].apply(lambda x: x.split(' ')[1] if pd.notnull(x) else None)
df
```

# Usuwanie kolumny 'Name'

```python
df.drop(columns=["Name"], inplace=True)
df
```

# Uzupełnienie brakujących danych

```python
df.info()
```

```python
df.sample(10)
```

```python
df.describe()
```

## Wiek zastąpię średnią

```python
df['Age'] = df['Age'].fillna(62.3)
```

## Adres zastąpię słowem "Nieznany"

```python
df['Address'] = df['Address'].fillna('Nieznany')
```

## Wzrost zastąpię średnią

```python
df['Height'] = df['Height'].fillna(174.5)
```

## Wagę zastąpię średnią

```python
df['Weight'] = df['Weight'].fillna(74.9)
```

## Imię zastąpię słowem "Nieznane"

```python
df['Imię'] = df['Imię'].fillna('Nieznane')
```

## Nazwisko zastąpię słowem "Nieznane"

```python
df['Nazwisko'] = df['Nazwisko'].fillna('Nieznane')
```

```python
df.sample(20)
```

```python
df.describe()
```

# Zmiana nazw kolumn

```python
df = df.rename(columns={
    "Age": "Wiek",
    "Address":"Adres",
    "Height":"Wzrost",
    "Weight":"Waga"})
df.sample()
```

# Wyliczenie BMI

```python
df["BMI"] = df["Waga"] / (df["Wzrost"] / 100) ** 2
df['BMI'] = df['BMI'].round(1)
df.sample()
```

# Sortowanie

```python
df.sort_values(by="BMI", ascending=False, inplace=True)
```

```python
df
```

# Zapisywanie

```python
df.to_csv('BMI.csv', index=False)
```
